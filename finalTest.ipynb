```python
# ---------------------------------------------
#  Import Libraries
# ---------------------------------------------
# description : Before running install prophet model and upgrade
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet

# ---------------------------------------------
#  Load & Prepare Data
# ---------------------------------------------
df = pd.read_csv('Data_Passangers.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df = df.sort_values(by='Date').fillna(0)

print(" Data Loaded & Cleaned")
display(df.head())
df.info()
display(df.describe())

# Numeric Columns
num_cols = df.select_dtypes(include=np.number).columns

# ---------------------------------------------
# Univariate Distribution Plots (Histogram)
# ---------------------------------------------
print("\n Distribution of Numeric Columns")
fig, axes = plt.subplots(nrows=(len(num_cols)+1)//2, ncols=2, figsize=(15, 5*((len(num_cols)+1)//2)))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df[col], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution: {col}')

for i in range(len(num_cols), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()

# ---------------------------------------------
#  Correlation Heatmap
# ---------------------------------------------
corr = df[num_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

#INSIGHT 1
# ---------------------------------------------
#  Scatter Plots: Rapid Route vs Others
# ---------------------------------------------
pairs = [('Local Route','Rapid Route'), ('Light Rail','Rapid Route')]
plt.figure(figsize=(12,5))

for i,(x,y) in enumerate(pairs,1):
    plt.subplot(1,2,i)
    sns.scatterplot(data=df,x=x,y=y)
    plt.title(f'{x} vs {y}')

plt.tight_layout()
plt.show()

# ---------------------------------------------
#  Boxplots (Outlier Detection Visual)
# ---------------------------------------------
for col in num_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot: {col}')
    plt.show()

# ---------------------------------------------
#  Outlier Analysis (IQR-based for "Other")
# ---------------------------------------------
Q1, Q3 = df['Other'].quantile([0.25, 0.75])
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

outliers = df[(df['Other'] < lower) | (df['Other'] > upper)]
print("\n Outlier Dates in 'Other':")
display(outliers['Date'])

plt.figure(figsize=(10,8))
sns.heatmap(outliers[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Outliers Only)")
plt.show()

#INSIGHT 2
# ---------------------------------------------
#  Monthly Trends
# ---------------------------------------------
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

monthly_avg = df.groupby(['Year','Month'])[num_cols].mean()
monthly_avg.index = pd.to_datetime(monthly_avg.index.map(lambda x: f"{x[0]}-{x[1]}"))

for col in num_cols:
    if col not in ['Year','Month']:
        monthly_avg[col].plot(figsize=(12,5), title=f"Monthly Average: {col}")
        plt.ylabel(col); plt.grid(); plt.show()

#INSIGHT 3
# ---------------------------------------------
#  Remove COVID Dip Period
# ---------------------------------------------
df_clean = df[(df['Date'] < '2020-05-01') | (df['Date'] > '2022-01-31')]
monthly_clean = monthly_avg[(monthly_avg.index < '2020-05-01') | (monthly_avg.index > '2022-01-31')]

for col in num_cols:
    if col not in ['Year','Month']:
        monthly_clean[col].plot(figsize=(12,5), title=f"{col} (Excluding May 2020‚ÄìJan 2022)")
        plt.ylabel(col); plt.grid(); plt.show()

#INSIGHT 4
# ---------------------------------------------
#  COVID Period Only Trends
# ---------------------------------------------
covid_df = df[(df['Date'] >= '2020-05-01') & (df['Date'] <= '2022-01-31')]
for col in num_cols:
    sns.lineplot(data=covid_df, x='Date', y=col)
    plt.title(f"{col} (COVID Period)"); plt.grid(); plt.show()

# ---------------------------------------------
#  Prophet Forecasting (Next 7 Days)
# ---------------------------------------------
forecast_cols = [c for c in num_cols if c not in ['Year','Month']]
forecasts = {}

print("\n Running Prophet Forecasts...")
for col in forecast_cols:
    data = df_clean[['Date',col]].rename(columns={'Date':'ds', col:'y'})
    model = Prophet()
    model.fit(data)

    future = model.make_future_dataframe(periods=7)
    fc = model.predict(future)
    forecasts[col] = fc[['ds','yhat','yhat_lower','yhat_upper']]

    print(f" Forecast Ready: {col}")

# Display Forecast Results
first = list(forecasts.keys())[0]
print("\n Sample Forecast:")
display(forecasts[first].head())

for col, fc in forecasts.items():
    print(f"\n 7-Day Forecast ‚Üí {col}")
    display(fc.tail(7))
```
